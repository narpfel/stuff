Dichteprofil entnommen von [hier](http://omniweb.gsfc.nasa.gov/vitmo/iri2012_vitmo.html).

```python
h, n_e = np.loadtxt("dichteprofil.csv", unpack=True)
```

```python
N = len(h)
```

```python
plt.plot(n_e, h)
plt.semilogx()
plt.grid()
plt.xlabel(r"$n_e / m^{-3}$")
plt.ylabel(r"$h / km$")
```

```python
plt.plot(n_e[:N // 3], h[:N // 3])
plt.semilogx()
plt.grid()
plt.xlabel(r"$n_e / m^{-3}$")
plt.ylabel(r"$h / km$")
```
